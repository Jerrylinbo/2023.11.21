```python
import cv2
import numpy as np
from urllib.request import urlopen
from matplotlib import pyplot as plt

def read_image_from_url(url):
    """Reads an image from a URL and returns it as a BGR numpy array."""
    response = urlopen(url)
    image = np.asarray(bytearray(response.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image

# Example image URL
url = 'https://p3-sign.toutiaoimg.com/tos-cn-i-6w9my0ksvp/950d86947b90486e89c74c23da077170~noop.image?_iz=58558&from=article.pc_detail&lk3s=953192f4&x-expires=1701465070&x-signature=KHT0wqupeBe25snIsBm%2BDA3Tv24%3D'  # URL

# Read the image from the URL
img_Nanami_bgr = read_image_from_url(url)

# Height, Width, and Three Channels
print('Image Size',img_Nanami_bgr.shape)

# Splitting the BGR image into its components
b, g, r = cv2.split(img_Nanami_bgr)

# Merging the components into an RGB image
img_Nanami_rgb = cv2.merge([r, g, b])

# Setting up the matplotlib plot
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(12, 4))

def f_imshow(ax, mat, title):
    ax.imshow(mat, cmap='gray')
    ax.set_title(title)
    ax.axis('off')

# Display each channel and the combined image
f_imshow(axs[0], img_Nanami_rgb, 'All Channels(RGB)')
f_imshow(axs[4], img_Nanami_bgr, 'Original Channels(BGR)' )
f_imshow(axs[1], r, 'Red Channel')
f_imshow(axs[2], g, 'Green Channel')
f_imshow(axs[3], b, 'Blue Channel')

# Adjust the layout of the subplots
plt.tight_layout(w_pad=0.5)
plt.show()
```
